```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import pandas as pd


service = Service(executable_path="chromedriver.exe")


driver = webdriver.Chrome(service=service)


driver.get("https://www.cityofsanrafael.org/major-planning-projects-2/")


matches = driver.find_elements(By.TAG_NAME, 'tr')
Project_Name = []
BriefProjectDescription = []
Floor_Area = []
NumberofUnits = []
NumberofBMRUnits = []
Applicant = []
Staff_Contact = []
status = []

# Print the text of each element
for match in matches:
    Project_Name.append(match.find_element(By.XPATH, './td[1]').text)
    project = match.find_element(By.XPATH, './td[2]').text
    BriefProjectDescription.append(project)
    print(project)

    Floor_Area.append(match.find_element(By.XPATH, './td[3]').text)
    NumberofUnits.append(match.find_element(By.XPATH, './td[4]').text)
    NumberofBMRUnits.append(match.find_element(By.XPATH, './td[5]').text)
    Applicant.append(match.find_element(By.XPATH, './td[6]').text)
    Staff_Contact.append(match.find_element(By.XPATH, './td[7]').text)
    status.append(match.find_element(By.XPATH, './td[8]').text)
    
 
    
    
    df=pd.DataFrame({'Project_Name':Project_Name,'BriefProjectDescription':BriefProjectDescription,'Floor_Area':Floor_Area,'NumberofUnits':NumberofUnits,'NumberofBMRUnits':NumberofBMRUnits,'Applicant':Applicant,'Staff_Contact':Staff_Contact,'status':status})
    df.to_csv('Project_data.csv',index=True)
    print(df)

   
```

```python
import pandas as pd

# Load the data from the CSV file
data = pd.read_csv('Project_data.csv')

# Remove the first row that contains meta descriptions
cleaned_data = data.drop(index=0)

# Standardize 'Floor_Area' by removing units and converting to numeric
cleaned_data['Floor_Area'] = cleaned_data['Floor_Area'].astype(str)
cleaned_data['Floor_Area'] = cleaned_data['Floor_Area'].replace({
    'sq.ft.': '', 'sqft': '', 'sf.': '', ',': '', ' ': '', 'sq.': '', 'ft.': ''
}, regex=True)
cleaned_data['Floor_Area'] = cleaned_data['Floor_Area'].str.extract(r'(\d+)')[0].astype(float)

# Convert 'NumberofUnits' and 'NumberofBMRUnits' to numeric, handling commas
cleaned_data['NumberofUnits'] = pd.to_numeric(cleaned_data['NumberofUnits'].replace(',', '', regex=True), errors='coerce')
cleaned_data['NumberofBMRUnits'] = pd.to_numeric(cleaned_data['NumberofBMRUnits'].replace(',', '', regex=True), errors='coerce')

# Display the cleaned and updated DataFrame
cleaned_data.head()
```

```python



from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import pandas as pd
import schedule
import time
import logging

# Set up logging configuration
logging.basicConfig(filename='web_scraping.log', level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

# Function to perform web scraping and update data
def update_data():
    try:
        # Set up the Chrome driver
        service = Service(executable_path="chromedriver.exe")
        driver = webdriver.Chrome(service=service)

        # Load the target webpage
        driver.get("https://www.cityofsanrafael.org/major-planning-projects-2/")

        # Find all table rows
        matches = driver.find_elements(By.TAG_NAME, 'tr')
        data = []

        # Extract the text of each element
        for match in matches:
            row_data = []
            try:
                # Append each cell's text to row_data
                cells = match.find_elements(By.TAG_NAME, 'td')
                if cells:  # Only proceed if there are cells in the row
                    for cell in cells:
                        row_data.append(cell.text)
                    # Append the complete row to data list
                    data.append(row_data)
            except Exception as e:
                logging.error(f"Error processing row: {e}")

        # Close the driver
        driver.quit()

        # Create a DataFrame from the collected data
        if data:
            columns = ['Project_Name', 'Brief_Project_Description', 'Floor_Area', 'Number_of_Units', 'Number_of_BMR_Units', 'Applicant', 'Staff_Contact', 'Status']
            df = pd.DataFrame(data, columns=columns)
            df.to_csv('Project_data.csv', index=False)
            logging.info("Data updated successfully.")
        else:
            logging.warning("No data found on the webpage.")
    except Exception as e:
        logging.error(f"Failed to update data: {e}")

# Schedule the data update task to run every day at 10:00 AM
schedule.every().day.at("10:00").do(update_data)


    
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
import pandas as pd

service = Service(executable_path="chromedriver.exe")
driver = webdriver.Chrome(service=service)
driver.get("https://www.tendersinfo.com/esearch/process?search_text=California+bids")

# Wait for the table rows to be present
WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.TAG_NAME, 'tr')))

matches = driver.find_elements(By.TAG_NAME, 'tr')

location = []
sector = []
date = []
summary = []
deadline = []

for match in matches:
    cells = match.find_elements(By.TAG_NAME, 'td')
    if len(cells) >= 5:  # Check if there are at least 5 cells
        location.append(cells[0].text)
        sector.append(cells[1].text)
        date.append(cells[2].text)
        
        summary_text = cells[3].text
        # Extract additional details like the organization name if present
        try:
            org_name = cells[3].find_element(By.TAG_NAME, 'b').text
            summary.append(summary_text + " (" + org_name + ")")
        except NoSuchElementException:
            summary.append(summary_text)  # Append without org name if <b> tag is missing
        
        deadline.append(cells[4].text)
    else:
        # If there aren't enough cells, append placeholders or log a warning
        location.append(None)
        sector.append(None)
        date.append(None)
        summary.append(None)
        deadline.append(None)
        print(f"Row skipped, not enough cells: {match.text}")

df = pd.DataFrame({
    'location': location,
    'sector': sector,
    'date': date,
    'summary': summary,
    'deadline': deadline
})
df.to_csv('Project2_data.csv', index=False)
print(df)
```
